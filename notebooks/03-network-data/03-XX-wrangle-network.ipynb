Create the card-deck network data

```python
import pandas as pd
import networkx as nx
```

```python
df = pd.read_feather('../../data/OTJ/deck_melt_card.feather')
df
```

```python
df_filtered = df[df['in_deck']]
len(df_filtered)
```

```python
B = nx.Graph()
```

```python
B.add_nodes_from(df_filtered['draft_idx'], bipartite=0)
B.add_nodes_from(df_filtered['card'], bipartite=1)
```

```python
B.add_edges_from(df_filtered[['draft_idx', 'card']].values)
```

```python
B.number_of_nodes(), B.number_of_edges()
```
